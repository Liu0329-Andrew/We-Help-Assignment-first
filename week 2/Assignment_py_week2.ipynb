```python
#Task 1
def find_and_print(messages, current_station):
    main_stations = [
            "Songshan", "Nanjing Sanmin", "Taipei Arena", "Nanjing Fuxing",
            "Songjiang Nanjing", "Zhongshan", "Beimen", "Ximen",
            "Xiaonanmen", "Chiang Kai-Shek Memorial Hall", "Guting", "Taipower Building",
            "Gongguan", "Wanlong", "Jingmei", "Dapinglin",
            "Qizhang", "Xindian City Hall", "Xindian"
        ]
    branch_station = ["Xiaobitan"]
    branch_connection = "Qizhang"
    connection_index = main_stations.index(branch_connection)
    
    current_index = 0
    if current_station in main_stations:
        current_index = main_stations.index(current_station)
        
    nearest_friend = None
    nearest_friend_distance = float('inf') 

    for friend, location in messages.items():
        friend_station_name = None
        for station in main_stations + branch_station:
            if station in location:
                friend_station_name = station
                break
                
        distance = float('inf')
        if friend_station_name in main_stations:
            if friend_station_name != current_station and current_station in branch_station:
                friend_station_index = main_stations.index(friend_station_name)
                distance = abs(friend_station_index - connection_index) + 1
            elif friend_station_name != current_station and current_station in main_stations:
                friend_station_index = main_stations.index(friend_station_name)
                distance = abs(friend_station_index - current_index)
            elif friend_station_name == current_station:
                distance = 0
        elif friend_station_name in branch_station:
            if friend_station_name != current_station:
                branch_distance = 1
                distance = abs(connection_index - current_index) + branch_distance
            elif friend_station_name == current_station:
                distance = 0
                
        if distance < nearest_friend_distance:
            nearest_friend_distance = distance
            nearest_friend = friend
            
    if nearest_friend:
        print(nearest_friend)
    

        
messages={
"Leslie":"I'm at home near Xiaobitan station.",
"Bob":"I'm at Ximen MRT station.",
"Mary":"I have a drink near Jingmei MRT station.",
"Copper":"I just saw a concert at Taipei Arena.",
"Vivian":"I'm at Xindian station waiting for you."
}

find_and_print(messages, "Wanlong") # print Mary
find_and_print(messages, "Songshan") # print Copper
find_and_print(messages, "Qizhang") # print Leslie
find_and_print(messages, "Ximen") # print Bob
find_and_print(messages, "Xindian City Hall") # print Vivian
find_and_print(messages, "Xiaobitan"); #print Vivian (我自己測試的情境)
```

```python
#Task 2
def book(consultants, hour, duration, criteria):
    if 'schedule' not in consultants[0]:
        for consultant in consultants:
            consultant['schedule'] = []
            
    start_time = hour
    end_time = hour + duration

    if criteria == "price":
        sorted_consultants = sorted(consultants, key=lambda x: x['price'])
    elif criteria == "rate":
        sorted_consultants = sorted(consultants, key=lambda x: x['rate'], reverse=True)
    else:
        print("Invalid criteria")
        return 

    for consultant in sorted_consultants:
        if is_available(consultant, start_time, end_time):
            if (start_time, end_time) not in consultant['schedule']:
                consultant['schedule'].append((start_time, end_time))
                print(consultant['name'])
                return 
                

    print("No Service")
    return 

def is_available(consultant, start, end):
    return all(end <= s[0] or start >= s[1] for s in consultant['schedule'])


consultants=[
{"name":"John", "rate":4.5, "price":1000},
{"name":"Bob", "rate":3, "price":1200},
{"name":"Jenny", "rate":3.8, "price":800}
]

book(consultants, 15, 1, "price"); # Jenny
book(consultants, 11, 2, "price"); # Jenny
book(consultants, 10, 2, "price"); # John
book(consultants, 20, 2, "rate"); # John
book(consultants, 11, 1, "rate"); # Bob
book(consultants, 11, 2, "rate"); # No Service
book(consultants, 14, 3, "price"); # John
```

```python
#Task 3
def func(*data):
    middle_name_counts = {}
    name_to_middle_name = {}

    for name in data:
        length = len(name)
        middle_name = None

        if length <= 3:
            middle_name = name[1]  # 第二個字
        elif length <= 5:
            middle_name = name[2]  # 第三個字
            
        if middle_name:
            name_to_middle_name[name] = middle_name
            middle_name_counts[middle_name] = middle_name_counts.get(middle_name, 0) + 1
            
    unique_names = []
    for name, middle_name in name_to_middle_name.items():
        if middle_name_counts[middle_name] == 1:
            unique_names.append(name)

    if not unique_names:
        print("沒有")
    else:
        print(", ".join(unique_names))




func("彭大牆", "陳王明雅", "吳明") # print 彭大牆
func("郭靜雅", "王立強", "郭林靜宜", "郭立恆", "林花花") # print 林花花
func("郭宣雅", "林靜宜", "郭宣恆", "林靜花") # print 沒有
func("郭宣雅", "夏曼藍波安", "郭宣恆") # print 夏曼藍波安
```

```python
#Task 4
def get_number(index):
    def generate_pattern(length):
        result = [0]
        a = 0
        while len(result) < length:
            b = a + 4
            c = b + 4
            d = c - 1
            result.extend([b, c, d])
            a = d
        return result[:length]

    print(generate_pattern(1000)[index])


get_number(1)  # print 4
get_number(5)  # print 15
get_number(10)  # print 25
get_number(30)  # print 70
```
